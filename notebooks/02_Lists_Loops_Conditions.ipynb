## 4. Conditions: HOT or NORMAL days
### ðŸ”¶ Task 4.1 â€“ Classify SST days

```python
threshold = 28.5
sst=[28.4,31.2,29.9,27.0,28.8,30.9,32.2]
for i in range(len(sst)):
    temp = sst[i]
    day = i + 1
    # TODO: write if condition here
    if temp > threshold:
      print("Day", day,":", "HOT")
    if temp < threshold:
      print("Day", day,":", "NORMAL")
```
